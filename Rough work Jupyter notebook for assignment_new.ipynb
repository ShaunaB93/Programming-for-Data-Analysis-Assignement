### Numpy.Random.Shuffle - 

As mentioned in the diagram above, this function modifies functions already in-place by shuffling its contents, in this case the contents will remain the same as the original array but the order will be altered. 

```python
# Example 1

arr = np.arange(20)
print(arr)
plt.plot(arr, "b")

np.random.shuffle(arr)
print(arr)
plt.plot(arr, "k")
```

*Function Output* - As can be seen in the examples above, this function re-arranges the output of print the numbers 0 to 20. This is represented in the plot as the blue continuous line as requested in the code. The original print statement is produced from numpy arranging the array between the range requested, in this case 0 and 20. While the .shuffle function linked with the random package re-arranges the original array each time it has run. This can ve visualisted by observing the variation in the black line plotted on the graph. Each time the code to produce this graph is re-ran, as the .shuffle function re-arranges the array this black line will shift to represent the re-arranged or shuffled order of the array.

### Numpy.Random.Permutation - 

As mentioned in the diagram above, this function randomly permutates a sequence or returns a permuted range. Similar to the .shuffle function explained above, the .permutation function will re-arrange the array either of the range provided upon each run of the code in a random manner.

```python
# Examples 1 and 2

arr2 = np.random.permutation(20)
print(arr2)
plt.plot(arr2, "k")

arr3 = np.random.permutation([1, 5, 15, 18, 20])
print(arr3)
plt.plot(arr3, "g")
```

*Function Output* - As can be seen in the examples above, this function permutates the order of either range provided to the function. 
The black line represents the output of the permutation function re-arranging the order of the range from 0 to 19 as the range provided to the function was up to, but not including, 20.
While the green output line provides a visualisation of the permutation function randomly re-ordering the list provided to the function each time the code is ran.
