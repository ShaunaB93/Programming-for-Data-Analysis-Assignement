# Overall purpose of numpy.random package in Python - 

The numpy.random package in python is useful for data analysis due to its ability to generate an array of random numbers using either the "rand(), randn() or randint()" functions.

It is particularly useful for hypothesis testing in data analysis due to its ability to provide a test sample of data. Sampling is known to be one of the most important useful factors in determining the accuracy of research during a project. As open source datasets can have a number of outliers or errors it can affect any testing as the results are reflective of the testing and the data used. By ruling out the errors which can be present in open datasets using this package it is more indicative or errors within the hypothesis when testing. This is a major advantage of having access to this package in python for data analysts during testing.

### Numpy.Random.Randint -
As mentioned in the diagram above this function provides random values from low(inclusive) to high(exclusive).

This function provides random numbers in an array of integers within the provided range. This function is very similar to the randint function previously mentioned with the addition of the low, high interval.

For example the function input below -

```python
# Example 1 - 

np.random.randint(5, size=8)
```

```python
# Example 2 -

np.random.randint(1, size=5)
```

```python
# Example 3 -

np.random.randint(1, 7, size=6)
```

Function Output - As can be seen in the examples above, this function produces the number of random integers outlned in the size argument within a range of 0, inclusive, and the 1st given argument within the function. The Example 2 above, confirms that if the argument range given is to 1 then all values in the array produced will be equal to 0 as there are no alternate available integer values for it to randomly choose.

### Numpy.Random.Random_integers -
As mentioned in the diagram above this function provides random values from low(inclusive) to high(exclusive) of type np.int.

This function produces similar results to the randint function. However, unlike randint if no lower argument is provided the values do not begin from 0 as the low value. Instead the default low value used is 1. As such this random_integers function can also be simulated using the randint function mentioned previously.

For example the function input below -

```python
# Example 1 - 

np.random.random_integers(10)
```

Function Output - While the above example does produce a result. The resulting error message advises this can be replaced using the np.random.randint(1, 10 + 1) function instead and produce a more stable result. This proves the simularities between both functions and that they can be used interchangably.
This random_integers function can be quite useful for testing where examples similar to rolling a dice etc. produce the required results to test a hypothesis.

### Numpy.random.random_sample / .random / .ranf / .sample

```python
# Example 2 - 
np.random.random_sample(5)
```

```python
# Example 3 - 
np.random.random_sample(10)
```
