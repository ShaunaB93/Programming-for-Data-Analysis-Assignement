## **_Overall purpose of the "numpy.random" package -_**

As can be seen from the title of the package the main purpose for this package within python is to generate arrays of random numbers. The use of random numbers in testing during analysis is widely used, especially as they can be formed in the shape of a dataset required, or with the required properties, for testing certain hypothesis, such as normal distribution etc. The ability to generate such datasets with the desired characteristics is very useful for testing various hypothesis where there is a degree of uncertainty without the additional effort required in collecting large datasets. But it must be noted that a prior knowledge of the system required to test the dataset is required before using it to test the hypothesis in question. For example, if you wished to test a hypothesis related to a characteristic of the population which is normally distributed, the array produced should have the same characteristic rather than generating a dataset of Poisson variables.*[1]* 

It should be noted when using such a package that there would be a subconscious bias in asking a person to select an array of random numbers there is a similar issue with generating random arrays using simulation. For this reason the numbers selected when using the “numpy.random” package are widely known as pseudorandom numbers instead. While the computer does not have a similar bias to that of which a person would have, they are still classed as pseudorandom in nature as they are selected by an algorithm with deterministic behaviour based on the seed of the random number generator. *[2]* This deterministic behaviour is based on the set of rules provided during the simulation, such as the size, characteristics of the dataset required, and the type of function used in the simulation. As the following examples will demonstrate there are a number of functions available for use within the “numpy.random” package to assist in producing the required type of dataset.


## **_Uses of the "Simple Random Data" functions -_** 

The functions included under the “Simple random data” heading within the numpy.random package can be utilised where a subset of the range provided or in-built is required where each value is equal probability of being chosen. It is useful where an unbiased representation of a group or range is required for testing. Selections made in this manner are useful when there is a larger population within a dataset as every element has an equal chance of being selected.
As can be seen in the following diagram some of the functions can have high and low values included as the range within the input for simulation while others have a built-in range, such as the random_sample function which includes a range of 0 up to, but excluding, 1. It can also be seen from this diagram that there are only slight variations in the functions and their output in this section of the numpy.random package.

There are also issues which such selections that analysts must be aware of. Although the selection is random using the simple data functions it can still lead to sampling errors in which the sample selected does not accurately reflect the dataset it is being used to represent. For this reason, and as mentioned previously, when using such functions, a prior knowledge of the system is required. For this reason, additional functions available within the “Distributions” section of the package, to be discussed below, may have a more specific distribution type which can be used compared to the “Simple Random Data” functions. For this reason the “Simple Random Data” functions are used as convenience functions and the “Distributions” functions are more widely utilised for testing with stricter conditions or distributions required. *[3]*

For example the numpy.random.random_int can be used to simulate a dice throw by using the following code -

```python
# Dice throw example, determine when rolling dice 20 times how many times the dice will land on the number 5.
# if outcome statement can be changed with any number between 1 and 6, the values on the face of a dice, to use this code to determine the probablility of the selected value appears when a dice is thrown 20 times. 

N = 20
dice = [np.random.randint(1, 7) for i in range(N)]
M = 0  # counter for successes: how many times we get 6 eyes
for outcome in dice:
    if outcome == 5:
        M += 1
print('Got five %d times out of %d' % (M, N))
print('Probability:', float(M)/N)
```

The functions included under the "Permutations" heading within the numpy.random simplifies randomly permutating or shuffling the data kept within a dataframe.

### Numpy.Random.Uniform function -

As suggested in the name of this function, and the explanation in the [documentation](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.uniform.html#numpy.random.uniform) , this function draws samples from a uniform distribution within the range specified based on a half open interval,meaning that within the interval each value is equally as likely to be selected. *[4]* There are similarities between this function, the random.rand() and random.randn() functions mentioned previously as each of the functions draw samples from a uniform distribution. However, with the .uniform function we can provide lower and upper range values for the sample generation.

From a statistics perspective, the uniform, or rectangular distribution, is a probability distribution as mentioned previously. It is also widely known to be a constant probability as it takes values between a specified range, between 0 and 1. This is common in statistics for random number generation. It is commonly used in probability calculations with the 0 end of the range indicating the event is unlikely to occur and while the end of the range, 1, indicates that the probability of the event in question is likely to occur. However, a uniform distribution can also be a discrete, meaning it will take a finite set of values. This specified range has two parameters - a, a minimum value, & b , the maximum value.

Similar to the example above in the alternate "Simple Random Data" function, the random.uniform function can also be used for simulating situations such as rolling a dice when used as a discrete distribution. It can be used in this case as each of the six sides of the dice have the same probability of occurring each time the dice is thrown. 

As the continuous uniform distribution is more commonly used for the random number generation the below examples will demonstrate this variation of this function. 

```python
N = 20
dice = [np.random.uniform(1, 7) for i in range(N)]
print("Values from the uniform distribution within the range of values from 20 dice rolls: \n", dice) 
plt.title('Dice roll probability using the numpy.random.uniform function')
plt.xlabel('Number on dice')
plt.ylabel('Number of occurances')
plt.hist(dice, bins=4)
```

```python
np.random.seed(5) #Setting the seed to provide reproducible results
sample1 = np.random.uniform(0,20,100) #Continuous distribution example
#All values within the sample range
np.all(sample1 >= 0)
np.all(sample1 < 20)
print("Values from the uniform distribution within the range of values 0 to 19 from 100 dice rolls: \n", dice) 
plt.title('Discrete distribution using numpy.random.uniform')
plt.xlabel('Lower and upper range values')
plt.ylabel('Number of occurances')
plt.hist(sample1, bins=20)
```

```python
uniform = np.random.uniform(low=0,high=5,size=100)
plt.plot(uniform)
plt.xlabel('X~U[0,1]')
plt.ylabel('Count')
plt.show()
plt.hist(uniform)
plt.xlabel('X~U[0,1]')
plt.ylabel('Count')
plt.show()
```

Non-technical explanation
If you pass n=1 to the Binomial distribution it is equivalent to the Bernoulli distribution. In this case the function could be thought of simulating coin flips. size=3 tells it to flip the coin three times and p=0.5 makes it a fair coin with equal probabilitiy of head (1) or tail (0).

The result of [1 0 0] means the coin came down once with head and twice with tail facing up. Technical explanation
Numpy implements random number generation in C. The source code for the Binomial distribution can be found here. Actually two different algorithms are implemented.

If n * p <= 30 it uses inverse transform sampling.
If n * p > 30 the BTPE algorithm of (Kachitvichyanukul and Schmeiser 1988) is used. (The publication is not freely available.)
I think both methods, but certainly the inverse transform sampling, depend on a random number generator to produce uniformly distributed random numbers. Numpy internally uses a Mersenne Twister pseudo random number generator. The uniform random numbers are then transformed into the desired distribution.

- A Binomially distributed random variable has two parameters n and p, and can be thought of as the distribution of the number of heads obtained when flipping a biased coin n times, where the probability of getting a head at each flip is p. (More formally it is a sum of independent Bernoulli random variables with parameter p).

For instance, if n=10 and p=0.5, one could simulate a draw from Bin(10, 0.5) by flipping a fair coin 10 times and summing the number of times that the coin lands heads.

In addition to the n and p parameters described above, np.random.binomial has an additional size parameter. If size=1, np.random.binomial computes a single draw from the Binomial distribution. If size=k for some integer k, k independent draws from the same Binomial distribution will be computed. size can also be an array of indices, in which case a whole np.array with the given size will be filled with independent draws from the Binomial distribution.

Note that the Binomial distribution is a generalisation of the Bernoulli distribution - in the case that n=1, Bin(n,p) has the same distribution as Ber(p).

### Numpy.Random.Binomial function -

Explanation - https://www.analyticsvidhya.com/blog/2017/09/6-probability-distributions-data-science/

As mentioned in the name this function draws samples randomly across a binomial distribution. Similar to the other functions in the numpy.random package it draws such samples within the specified parameters given in the input. The binomial distribution has uses in experiments that have on of two different types of outcomes. Such outcomes can be yes or no, true or false, success or failure which makes it an important distribution in statistical analysis. It is also a discrete probability. The parameters filled for this function are as follows - 

n -> number of trials, this is a finite number
p -> probability of success where n is greater than 0 and p is within the interval [0,1). This is due to the fact that probability, as discussed, is given as a value between 0 and 1 where 0 indicates the event is unlikely to occur and 1 indicates the event is likely to occur.

The generally known characteristics of a binomial distribution are - 
1) The event is repeated the given n parameter number of times and the repetitions must be identical.
2) Each trial is independent of the other events.
3) There are only two possible outcomes of the event, as described previously.

The following histogram will provide a general overview of the shape of a binomial distribution visualisation in statistics - 

```python
np.random.seed(5) #Provide seed to lead to a reproducible outcome for discussion
binomial1 = np.random.binomial(10, 0.5, 20) #Represents a Bernoulli trial completed 10 times over 20 independent trials
M= 0 #counter for successes: how many times probability is above 0.5 and indicates success
for outcome in binomial1:
    if outcome > 0.5:
        M += 1
    elif outcome < 0.5:
        M += 0
print("Outcome array of success being the result in 10 trials containing 20 independent tests: \n", binomial1) 
plt.title('Coin flip outcome using Bernoulli function in the numpy.random package')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.hist(binomial1)
```

```python
np.random.seed(5) #Set the seed for reproducibility
binomial2= np.random.binomial(10, 0.25, 20) #using binomial function with 0.25 or 25% probability of likelyhood for comparison.
binomial2= np.random.binomial(10, 0.75, 20) #using binomial function with 0.75 or 75% probability of likelyhood for comparison.
print('Outcome of binomial function with 0.5 probability of being randomly generated: ', binomial1)
print('Outcome of binomial function with 0.25 probability of being randomly generated: ', binomial2)
print('Outcome of binomial function with 0.75 probability of being randomly generated: ', binomial3)
plt.plot(binomial1, 'g-', label='0.5 probability')
plt.plot(binomial2, 'k-', label='0.25 probability')
plt.plot(binomial3, 'b-', label='0.75 probability')
plt.xlabel('Number of successes')
plt.ylabel('Probability of successes')
plt.legend()
plt.show()
```

## **_Uses of seeds in generating pseudorandom numbers -_** 

The seed function within the numpy.random package is the parameter which is used to start the sequence in generating the pseudorandom numbers requested. This function is deterministic, as mentioned previously, which means that if the same seed is used the function would produced the same sequence of numbers. *[8]*

This function takes an integer value argument, similar to the previous discussed values.
