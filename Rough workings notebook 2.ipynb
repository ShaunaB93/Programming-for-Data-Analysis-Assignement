## **_Overall purpose of the "numpy.random" package -_**

As can be seen from the title of the package the main purpose for this package within python is to generate arrays of random numbers. The use of random numbers in testing during analysis is widely used, especially as they can be formed in the shape of a dataset required, or with the required properties, for testing certain hypothesis, such as normal distribution etc. The ability to generate such datasets with the desired characteristics is very useful for testing various hypothesis without the additional effort required in collecting large datasets. But it must be noted that a prior knowledge of the system required to test the dataset is required before using it to test the hypothesis in question. For example, if you wished to test a hypothesis related to a characteristic of the population which is normally distributed, the array produced should have the same characteristic rather than generating a dataset of Poisson or discrete variables.*[1]* 

It should be noted when using such a package that there would be a subconscious bias in asking a person to select an array of random numbers there is a similar issue with generating random arrays using simulation. For this reason the numbers selected when using the “numpy.random” package are widely known as pseudorandom numbers instead. While the computer does not have a similar bias to that of which a person would have, they are still classed as pseudorandom in nature as they are selected by an algorithm with deterministic behaviour based on the seed of the random number generator. *[2]* This deterministic behaviour is based on the set of rules provided during the simulation, such as the size, characteristics of the dataset required, and the type of function used in the simulation. As the following examples will demonstrate there are a number of functions available for use within the “numpy.random” package to assist in producing the required type of dataset.


## **_Uses of the "Simple Random Data" functions -_** 

The functions included under the “Simple random data” heading within the numpy.random package can be utilised where a subset of the range provided or in-built is required where each value is equal probability of being chosen. It is useful where an unbiased representation of a group or range is required for testing. Selections made in this manner are useful when there is a larger population within a dataset as every element has an equal chance of being selected.
As can be seen in the following diagram some of the functions can have high and low values included as the range within the input for simulation while others have a built-in range, such as the random_sample function which includes a range of 0 up to, but excluding, 1. It can also be seen from this diagram that there are only slight variations in the functions and their output in this section of the numpy.random package.

There are also issues which such selections that analysts must be aware of. Although the selection is random using the simple data functions it can still lead to sampling errors in which the sample selected does not accurately reflect the dataset it is being used to represent. For this reason, and as mentioned previously, when using such functions, a prior knowledge of the system is required. For this reason, additional functions available within the “Distributions” section of the package, to be discussed below, may have a more specific distribution type which can be used compared to the “Simple Random Data” functions. For this reason the “Simple Random Data” functions are used as convenience functions and the “Distributions” functions are more widely utilised for testing with stricter conditions or distributions required. *[3]*

### Numpy.Random.Rand() & Numpy.Random.Randn() functions- 

As mentioned in the diagram above the Numpy.Random.Rand() function provides random values in the requested shape in an array on the interval [0,1). While the Numpy.Random.Randn() function provides random values with random floats sampled from a univariate “normal” (Gaussian) distribution of mean 0 and variance 1. *[4]*

The ".rand" and ".randn" functions are quite widely used as they produce arrays across the normal and uniform distributions, respectively, and can be used to obtain multi dimensional arrays also. *[5]*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Example 1
# The below code snippet outputs 5 arrays each with 3 elements over the [0, 1) interval -

np.random.rand(5,3)
```

```python
# Example 2
# The below code produces both a plot and histogram for a random 5 element array produced using the .rand function. 

arr1 = np.random.rand(5) 
print("Array of values produced using the random.rand function: \n", arr1) 
  
plt.plot(arr1)
plt.title(".Rand Function Plot")
plt.ylabel("Value")
plt.show()

plt.hist(arr1)
plt.title("Historgram")
plt.xlabel("Value")
plt.show() 
```

```python
# Example 3 - 
# The below code snippet outputs 2 arrays each with 4 elements.
# Range differs from the .rand function above.

np.random.randn(2, 4)
```

```python
# Example 4 -
# The below code produces both a plot and histogram for arrays produced using the .randn function.

arr2 = np.random.randn(2, 4) 
print("Array of values produced using the random.randn function: \n", arr2) 
  
plt.plot(arr2)
plt.title(".Randn Function Plot")
plt.ylabel("Value")
plt.show()


plt.hist(arr2)
plt.title(".Randn Function Plot")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show() 
```

## **_Uses of the "Permutations" functions -_** 

The functions included under the "Permutations" heading within the numpy.random simplifies randomly permutating or shuffling the data kept within a dataframe. 

This can be quite useful in machine learning instances when dividing a dataset in two to have a training and testing dataset to test a model. For example, in the instance of the famous [Iris dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set), the dataset itself is ordered in a way that the categories of flowers are grouped. It can be useful to use the numpy.random.permutations functions to shuffle this dataset if it is being used for a machine learning purpose so that both the training and testing datasets contain randomly ordered variations of the dataset. If this is not completed, and for example the analyst selects the last 20 rows of information to be the testing dataset, the model will only be tested providing information for one selected category of flower. This would not be a good testing of the machine learning module used.

### Examples of the "Permutations" functions from numpy.random package - 

 *[4]*

### Numpy.Random.Shuffle() & Numpy.Random.Permutation() functions- 


```python
# Example 5

arr3 = np.arange(20)
print("Array of values arranged for range provided \n", arr3) 
plt.plot(arr3, "b")

np.random.shuffle(arr3)
print("Shuffled range of valyes within the range provided: \n", arr3) 
plt.plot(arr3, "k")
```

```python
# Example 6

arr4 = np.random.permutation(20)
print("Permuted range of values within range provided: \n", arr4)
plt.plot(arr4, "k")

arr4 = np.random.permutation([1, 5, 15, 18, 20])
print("Permuted range of values within range provided: \n", arr4) 
plt.plot(arr4, "g")
```
