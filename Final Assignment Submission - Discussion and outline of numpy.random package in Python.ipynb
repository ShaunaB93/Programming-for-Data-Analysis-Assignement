## **_Overall purpose of the "numpy.random" package -_**

As per the title of the package the main purpose of this package within python is to generate an array of random numbers. The use of random numbers in testing during analysis is widely used, especially as they can be formed in the shape of a dataset required, or with the required properties, for testing certain hypothesis, such as normal distribution etc. The ability to generate such datasets, with the desired characteristics, is very useful for testing various hypothesis where there is a degree of uncertainty without the additional effort required in collecting large datasets. However, it should be noted that a prior knowledge of the system required to test the dataset is essential before using it to test the hypothesis in question. For example, if you wished to test a hypothesis related to a characteristic of the population which is normally distributed, the array produced should have the same characteristic rather than generating a dataset of Poisson variables.*[1]* 

In addition, it should be noted when using such a package that there would be a subconscious bias in asking a person to select an array of random numbers there is a similar issue with generating random arrays using simulation. For this reason the numbers selected when using the “numpy.random” package are widely known as pseudorandom numbers instead. While the computer does not have a similar bias to that of which a person would have, they are still classed as pseudorandom in nature as they are selected by an algorithm with deterministic behaviour based on the seed of the random number generator. *[2]* This deterministic behaviour is based on the set of rules provided during the simulation, such as the size, characteristics of the dataset required, and the type of function used in the simulation. The following examples demonstrate a number of functions available for use within the “numpy.random” package to assist in producing the required type of dataset.

### Hypergeometric distribution –
#### Profile – 
#### Shape of the distribution –


 
*[15]*
The hypergeometric distribution simulates random number generation by drawing values as a discrete probability without replacement from a finite population size where the outcome can be of two variations, such as success and failure. This has its similarities to the binomial distribution mentioned previously, while the binomial distribution generates values with replacement the hypergeometric distribution without replacement. However, due to the similarities the binomial distribution can be a representative, or approximation, of the hypergeometric distribution once the sample is 5% or less of the population range. This distribution differs from those mentioned previously as the values drawn are removed from future trials rather than still having the possibility to be selected.

#### Parameters for the function in the numpy.random.hypergeometric package –
As mentioned in the scipy documentation for the hypergeometric function the following are the required parameters when utilising this function – 
numpy.random.hypergeometric(ngood, nbad, nsample, size=None) *[16]*
From utilising the above, the randomly generated dataset is drawn from a hypergeometric distribution using the *ngood* parameter, number of ways of making a good selection, *nbad*, number of ways of making a bad selection, *nsample*, the number of elements, and the optional size parameter, which gives the output a specific shape. It must be noted that the *nsample* function must be less than or equal to the sum of the *ngood* and *nbad* values in order for the probability to be correctly calculated.
#### Uses for the function in the numpy.random.hypergeometric package –
Such sampling from a population can be very useful in many instances in data analysis. One example used is for determining the probability of obtaining defect or outliers in a batch of products using this distribution without replacement. Such as in a sample size of 50 out of 100 produced products how likely is it when 10 samples are drawn a defective good will be selected. The below example utilises the same principle when drawing different coloured balls from a draw and the probability of drawing each of the colours.


```python
np.random.seed(5) #set seed for reproducibility of result
nred = 20 # number of red balls in selection
ngreen = 8 # number of green balls in selection
nblue = 22 # number of blue balls in selection

m = 15 # sample size

red = np.random.hypergeometric(nred, ngreen + nblue, m)
green = np.random.hypergeometric(ngreen, nblue, m - red)
blue = m - (red + green)

print("When randomly selecting 15 of the 20 balls from the sample the number of red balls selected is: %2i" % red)
print("When randomly selecting 15 of the 20 balls from the sample the number of red balls selected is: %2i" % green)
print("When randomly selecting 15 of the 20 balls from the sample the number of red balls selected is: %2i" % blue)
```

### Poisson distribution –
#### Profile – 
#### Shape of the distribution –


 
*[17]*
The Poisson distribution is used to randomly simulate the probability of certain events when you already know how frequent an event has occurred previously. This distribution works over a fixed interval of time rather than the intervals mentioned previously. The distribution takes its name as it predicts the probability of an event based on completing or simulating independent Poisson experiments. A Poisson experiment is one the matches the following characteristics:
1)	The outcome is classed as a success or failure, similar to the binomial distribution 
2)	The average number of successes over a specified time interval is already known.
3)	The probability of success is proportional to the size of the period in question such as a period of time or volume.

#### Parameters for the function in the numpy.random.poisson package –
As mentioned in the scipy documentation for the poisson function the following are the required parameters when utilising this function – 
numpy.random.poisson(lam=1.0, size=None)
From this outline of the function, in order to simulate a dataset of poisson distribution two parameters are required. The lam parameter is the expectation of the interval in question or the average number of events over the known interval. While the size parameter provides the output shape, such as the number of samples drawn. Due to its similarities to the binomial function it is said to be the limit of the binomial distribution for trials with a large N sample size.
*[18]*
#### Uses for the function in the numpy.random.poisson package –
As the Poisson distribution can be used to predict the likelihood of an event over a specified time frame is very useful for hypothesis testing across data analytics in many different areas. One such example is to predict, based on a population size in a city, how many car crashes are likely to happen over the next period of time in question. For example, the below demonstrates random datasets where the previous average of events, such as crashes meant previously, is utilised to predict the number over the next period –

####  Relationship between the Binomial and Hypergeometric distributions – 

The relationship between the Binomial and Hypergeometric relationship has been generally described in the Hypergeometric profile above. The similarities or relationship are due to the manner in which both distributions randomly draw or simulate their results. Both distributions are known to be discrete probability distributions that represent the outcomes of randomly simulating Bernoulli trials. Although both distributions draw random samples from finite population within this package the main difference is that when the Binomial function is used the outcomes are replaced for each trial, meaning the same outcome has the same probability each time a trial is completed. However, when using the Hypergeometric distribution the values are randomly simulated without replacement. This can be described or better seen using the drawing of the coloured balls example simulated in the Hypergeometric profile.
In this case if a Binomial distribution is used in the numpy.random package for this situation the ball drawn in each trial would be placed back in to the possible selections for the next trial, therefore the probability of it being drawn does not change between trials. If the hypergeometric random distribution is used in this same situation, the next trials are completed without the ball from the last drawn being placed back in to the simulation and cannot be selected again. In this case the probability each time a trial is completed without replacement changes. Another of the differences between the functions is that the trials are not dependent on each other when using the Binomial distribution to simulate the drawing of samples. But the Hypergeometric trials are dependent due to the lack of replacement in such trials.
 Due to the similarities in the conditions which need to be met for using both distributions the accuracy is mainly determined, and the functions interchanged for uses depends on whether the relationship in the trial being completed is independent or not. For example, if the trial was to flip a two-sided coin the second side in theory can’t be replaced and therefore using a hypergeometric distribution simulation in this case would not be accurate as when repeating this trial the results are independent of each other and therefore a binomial distribution simulate would have increased accuracy.

The seed function within the numpy.random package is the parameter which is used to start the sequence in generating the pseudorandom numbers requested. This function is deterministic, as mentioned previously, which means that if the same seed is used the function would produced the same sequence of numbers. *[8]*

This function takes an integer value argument, similar to the previous discussed values. But if not specified it does have a default value using a global random seed which is the current system time in milliseconds from epoch(1970). *[8]* Which must be considered each time a testing is completed using the numpy.random package, as if an array of random numbers is required that can be reproduced it can be useful to set the seed when writing the code for analysis. If a seed can be determined it is also possible to determine the manner in which it is pulling the data "randomly" from the dataset. 

As the seed can be set or determined and there are characteristics given to most functions contained within this package this is the reason the numbers generated are not truly random and are known as pseudorandom numbers.

Each of the below variations of the seed function can be used to set a seed for your analysis as required.

### Numpy.Random.RandomState function - 

The random.randon state function provides multiple methods of generating random numbers which can be drawn from the variety of probability distributions available within the numpy.random package mentioned above. There are 3 variations on this function depending on the argument in the *size* argument of the function - 

1) _size = None_

    a single value is generated and returned
    
2) _size = integer_

    a one dimensional array is generated and returned

3) _size = tuple_

    an array of the requested shape is generated and returned



### Numpy.Random.Seed function - 

This function is a follow on from the RandomState function in generating seeds as it is called once the RandomState function is initialised. If a re-seed of the pseudorandom number generator is required this function can also be re-called.
